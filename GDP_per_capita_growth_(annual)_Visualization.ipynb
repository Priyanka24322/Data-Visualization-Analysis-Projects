```python
import pandas as pd
from tqdm.notebook import tqdm


# Loading the data
filepath = 'https://raw.githubusercontent.com/Priyanka24322/Data-Visualization-Analysis-Projects/refs/heads/main/Datasets/GDP_Per_capita_Growth_anually_.csv'
df = pd.read_csv(filepath)

df.head(10)
```

```python
import matplotlib.pyplot as plt

# Replacing ".." with NaN
# Convert the year columns to numeric, errors='coerce'
for col in df.columns[4:]:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Drop the non-numeric columns
numeric_df = df.drop(columns=['Series Name', 'Series Code', 'Country Name', 'Country Code'])

# Plotting the distribution of GDP per capita growth for all years
plt.figure(figsize=(15, 10))
for i, column in enumerate(numeric_df.columns):
    plt.subplot(3, 4, i+1)
    numeric_df[column].hist(bins=20)
    plt.title('Distribution in ' + column[:4])
    plt.xlabel('GDP per capita growth (%)')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

The plots above show the distribution of GDP per capita growth rates for the years where data is available. Each subplot represents a different year, with the x-axis showing the growth rate percentage and the y-axis indicating the frequency of countries within each growth rate bin. The histograms provide a quick way to assess the spread and central tendency of GDP growth rates across the sampled countries for each year.

```python
import plotly.express as px

india_data = df[df['Country Name'] == 'India']

india_data_melted = pd.melt(india_data, id_vars=['Country Name'],
                            value_vars=['1990 [YR1990]', '2000 [YR2000]', '2013 [YR2013]',
                                        '2014 [YR2014]', '2015 [YR2015]', '2016 [YR2016]',
                                        '2017 [YR2017]', '2018 [YR2018]', '2019 [YR2019]',
                                        '2020 [YR2020]', '2021 [YR2021]', '2022 [YR2022]'],
                            var_name='Year', value_name='GDP Growth Rate')

fig = px.line(india_data_melted, x='Year', y='GDP Growth Rate',
              title='GDP Per Capita Growth Over the Years for India',
              labels={'GDP Growth Rate': 'GDP Per Capita Growth Rate'})
fig.show()
```

```python
import seaborn as sns

melted_df = df.melt(id_vars=['Country Name', 'Country Code'],
                     var_name='Year', value_name='GDP Growth (%)',
                     value_vars=df.columns[4:])


melted_df = melted_df.dropna(subset=['GDP Growth (%)'])
```

```python
# Convert 'Year' to a string
melted_df['Year'] = melted_df['Year'].str.extract('(\d{4})')

# Convert 'Year' to integer
melted_df['Year'] = melted_df['Year'].astype(int)
```

```python
# Plotting GDP Growth over the years for all countries
plt.figure(figsize=(12, 6))
sns.lineplot(data=melted_df, x='Year', y='GDP Growth (%)', estimator='mean', ci=None)
plt.title('Average GDP per Capita Growth Over Years')
plt.xlabel('Year')
plt.ylabel('Average GDP Growth (%)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

- By looking at the above visualization we can see that there we a sudden dip in the Average GDP Growth in the year 2020, which indicates that there was a severe impact on economy suring the corona period.

- Looking at the pervious trend on year before 2000 it can be seen that there was a growing trend in the GDP growth, which might have happend because of robust urbanization in the Asian countries like China, Japan, South Korea etc.

- But the dropdown indicates that there was a decrease in the GDP percentage after the year 2000, it may have happend because of the recession that was caused in Japan and some other asian countries, which was in a down fall till 2012-2013, and this might explain the impact of the Great Recession of 2008 on the Global GDP average.

- After the year 2021 the world has recovered from the impacts of Lockdown and is in the progress of positve growth.


```python
subset_df = df.replace('..', pd.NA).dropna(subset=['2017 [YR2017]', '2018 [YR2018]', '2019 [YR2019]', '2020 [YR2020]', '2021 [YR2021]'])

subset_df = subset_df[['Country Name', '2017 [YR2017]', '2018 [YR2018]', '2019 [YR2019]', '2020 [YR2020]', '2021 [YR2021]']]

melted_df = subset_df.melt(id_vars=['Country Name'], var_name='Year', value_name='GDP Growth')

melted_df['GDP Growth'] = pd.to_numeric(melted_df['GDP Growth'], errors='coerce')

plt.figure(figsize=(12, 6))
sns.boxplot(x='Year', y='GDP Growth', data=melted_df)
plt.title('Distribution of GDP Per Capita Growth by Year')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
sorted_df = subset_df.sort_values(by='2021 [YR2021]', ascending=False)

# Get the top 5 countries
top_countries = sorted_df.head(5)

# Get the bottom 5 countries
bottom_countries = sorted_df.tail(5)

print('Top 5 countries by GDP per capita growth in 2021:')
print(top_countries)
print('\nBottom 5 countries by GDP per capita growth in 2021:')
print(bottom_countries)
```

```python
years = ['2017 [YR2017]', '2018 [YR2018]', '2019 [YR2019]', '2020 [YR2020]', '2021 [YR2021]']

for year in years:
    df[year] = df[year].replace('..', pd.NA).apply(pd.to_numeric, errors='coerce')

filtered_df = df.dropna(subset=years)

top_countries = filtered_df.nlargest(5, '2021 [YR2021]')['Country Name']
bottom_countries = filtered_df.nsmallest(5, '2021 [YR2021]')['Country Name']

filtered_df = filtered_df[filtered_df['Country Name'].isin(top_countries) | filtered_df['Country Name'].isin(bottom_countries)]

filtered_df.set_index('Country Name', inplace=True)

transposed_df = filtered_df[years].transpose()

plt.figure(figsize=(14, 7))
for country in top_countries:
    plt.plot(transposed_df.index, transposed_df[country], marker='o', label=country)
for country in bottom_countries:
    plt.plot(transposed_df.index, transposed_df[country], marker='o', label=country)

plt.title('GDP Per Capita Growth Trend (2017-2021)')
plt.xlabel('Year')
plt.ylabel('GDP Per Capita Growth (%)')
plt.xticks(rotation= 0)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
sorted_df = subset_df.sort_values(by='2021 [YR2021]', ascending=False)

# Get the top 40 countries
top_countries = sorted_df.head(40)

plt.figure(figsize=(12, 8))
plt.bar(top_countries['Country Name'], top_countries['2021 [YR2021]'], label='2021')
plt.xlabel('Country')
plt.ylabel('GDP per Capita Growth')
plt.title('Comparison of GDP Per Capita Growth Among Top 30 Countries (Stacked Bar Chart)')
plt.legend()
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
import squarify

sizes = top_countries['2021 [YR2021]'].values
labels = top_countries['Country Name'].values

# Draw the tree map
plt.figure(figsize=(12, 8))
squarify.plot(sizes=sizes, label=labels, alpha=.9)
plt.axis('off')
plt.title('Tree Map of GDP Per Capita Growth in 2021 (Top 40 countries)')
plt.show()
```

```python
# Draw the tree map using Plotly
fig = px.treemap(top_countries, path=['Country Name'], values='2021 [YR2021]',
                 title='Tree Map of GDP Per Capita Growth in 2021 (Top 40 Countries)',
                 labels={'2021 [YR2021]': 'GDP Per Capita Growth'})

fig.update_layout(margin=dict(t=50, l=0, r=0, b=0))

fig.show()
```
